Students will implement natural join using two strategies:

- Compute Cartesian product → selection → projection.
- Use a hash-join algorithm.

```python
# [THIS IS READ-ONLY]
import csv
import time
import pandas as pd

def read_csv(csvfile):
    with open(csvfile, 'r') as f:
        rdr = csv.reader(f)
        columns = tuple(next(rdr))
        rows = list(map(tuple, rdr))
        return columns, rows
```

```python
# [YOUR WORK HERE]
# @workUnit

# return columns as a list[str], and rows as a list[tuple]

def natural_join(csvfile1, csvfile2):
    col1, rows1 = read_csv(csvfile1)
    col2, rows2 = read_csv(csvfile2)

    # Find common columns
    common_columns = [c for c in col1 if c in col2]
    
    all_columns = list(col1) + [c for c in col2 if c not in common_columns]

    # Create hash table
    join_indices1 = [col1.index(c) for c in common_columns]
    hash_table = {}
    for r1 in rows1:
        key = tuple(r1[i] for i in join_indices1)
        hash_table.setdefault(key, []).append(r1)
        
    # Join rows 1 with rows2
    join_indices2 = [col2.index(c) for c in common_columns]
    joined_rows = []
    for r2 in rows2:
        key = tuple(r2[i] for i in join_indices2)
        if key in hash_table:
            for r1 in hash_table[key]:
                merge = r1 + tuple(r2[j] for j in range(len(r2)) if col2[j] not in common_columns)
                joined_rows.append(merge)
                
    return tuple(all_columns), joined_rows
```

```python
# [THIS IS READ-ONLY]
# @check
# @title: small join

c, rs = natural_join('r0_small.csv', 'r1_small.csv')
c_answer, rs_answer = read_csv('answer_small.csv')

rs = set(rs)
rs_answer = set(rs_answer)

assert c == c_answer

print("Unexpected tuples:")
df = pd.DataFrame(data=list(rs - rs_answer), columns=c)
print(df)

print("===")
print("Missing tuples:")
df = pd.DataFrame(data=list(rs_answer - rs), columns=c)
print(df)
```

```python
# [THIS IS READ-ONLY]
# @check
# @title: large join

c, rs = natural_join('r0_large.csv', 'r1_large.csv')
c_answer, rs_answer = read_csv('answer_large.csv')

rs = set(rs)
rs_answer = set(rs_answer)

assert c == c_answer

print("Unexpected tuples:")
df = pd.DataFrame(data=list(rs - rs_answer), columns=c)
print(df)

print("===")
print("Missing tuples:")
df = pd.DataFrame(data=list(rs_answer - rs), columns=c)
print(df)
```
