# Learning about database API in Python

In this section, you will have to complete the following tasks by learning the details on your own (with the help of this lab).

````python
# [THIS IS READ-ONLY]
#
# We will use the Psycopg2 library which is a postgres database connectivity
# library for Python.
#

import psycopg2 as pg
````

````python
# [YOUR WORK HERE]
# @workUnit

# First we need to construct a database connection.  All database activities
# happen via the connection object.

# You need to learn on how to specify a connection string in the form of "postgres://..."
# The database to connect is called "jovyan".

# Optionally, you can provide the follow: 
# - the database server "127.0.1".
# - port: 5432
# - user: jovyan
#
# These optionally data need to packed into the connection string.

conn = pg.connect("postgresql://jovyan@127.0.0.1:5432/jovyan")
````

Next, we need to create a cursor object.  A cursor is an object that scans through results
of a SQL query.

````python
# [YOUR WORK HERE]
# @workUnit

# create a cursor object from the `conn` object.

cur = conn.cursor()
````

Now, we can use the cursor to execute a SQL query.

````python
# [THIS IS READ-ONLY]
sql = "select * from artist"
````

````python
# [YOUR WORK HERE]
# @workUnit

#
# use the execute method of the cursor object to run the query.
#

cur.execute("SELECT * FROM artist;")
````

````python
# [YOUR WORK HERE]
# @workUnit

# cursor provides a number of fetch methods.
# use `fetchone()` method to fetch the first row.

row = cur.fetchone()
````

````python
# [THIS IS READ-ONLY]
# @check
# @title: first row

print(row)
````
