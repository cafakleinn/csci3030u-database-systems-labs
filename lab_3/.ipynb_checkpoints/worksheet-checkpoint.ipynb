## Adding constraints

Your job is to alter the tables to create the following data integrity constraints.  The `ALTER TABLE` statements should
all go into a single `add_constraints.sql` file.

| **Constraint Name**                     | **Description**                                                                                 |
| --------------------------------------- | ----------------------------------------------------------------------------------------------- |
| `authors_pkey`                          | Primary key on `authors.author_id`, ensures each author has a unique ID.                        |
| `posts_pkey`                            | Primary key on `posts.post_id`, ensures each post has a unique ID.                              |
| `tags_pkey`                             | Primary key on `tags.tag_id`, ensures each tag has a unique ID.                                 |
| `post_tags_pkey`                        | Composite primary key on `(post_id, tag_id)`, prevents duplicate tag links.                     |
| `authors_email_key`                     | Unique constraint on `authors.email`, ensures each email appears once.                          |
| `tags_name_key`                         | Unique constraint on `tags.name`, ensures each tag name is unique.                              |
| `posts_author_id_fkey`                  | Foreign key from `posts.author_id`  `authors.author_id`, deletes posts when author is deleted. |
| `post_tags_post_id_fkey`                | Foreign key from `post_tags.post_id`  `posts.post_id`, deletes links if post is deleted.       |
| `post_tags_tag_id_fkey`                 | Foreign key from `post_tags.tag_id`  `tags.tag_id`, deletes links if tag is deleted.           |
| `authors_name_not_null` *(implicit)*    | Enforces `authors.name` must not be NULL.                                                       |
| `authors_email_not_null` *(implicit)*   | Enforces `authors.email` must not be NULL.                                                      |
| `posts_author_id_not_null` *(implicit)* | Enforces `posts.author_id` must not be NULL.                                                    |
| `posts_title_not_null` *(implicit)*     | Enforces `posts.title` must not be NULL.                                                        |
| `tags_name_not_null` *(implicit)*       | Enforces `tags.name` must not be NULL.                                                          |
